```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset_bank = pd.read_csv("bankmarketing.csv")
dataset_bank.head(10)
```

```python
dataset_bank.columns
```

dari data di bawah ini durasi pada hari kerja terbanyak berada di hari senin, 
dan durasi pada hari kerja paling sedikit berada pada hari rabu

```python
x = dataset_bank['day_of_week']
y = dataset_bank['duration']
nct = "#bfff00"
plt.bar(x,y, align = 'center', width = 0.7, color = nct)
plt.title('Durasi PerHari')
plt.xlabel('DAY OF WEEK')
plt.ylabel('DURATION')
plt.show()
```

dari data scatterplot di bawah ini, pekerjaan retired paling banyak dari berbagai usia pada data, dengan rata-rata usia pekerja berada pada usia produktif.
Sedangkan pekerjaan hanya tergolong sedikit dari berbagai usia pada data adalah student, dengan rata-rata usia pada usia muda

```python
maroon = "#833141"
sns.scatterplot(data=dataset_bank, x='age', y='job', color = maroon)
plt.title('Pekerjaan Berdasarkan Umur')
plt.xlabel('AGE')
plt.ylabel('PEKERJAAN')
plt.show()
```

```python
dataset_bank = dataset_bank.dropna(subset=['job'])
```

```python
data_pie = dataset_bank['job'].value_counts().rename_axis('job').reset_index(name='age')
```

```python
data dibawah ini menunjukkan diagram pie dengan pekerjaan paling banyak dari data bank adalah admin dengan 25.3% 
dan pekerjaan paling sedikit adalah uknown atau yang lainnya dengan 0.8%
```

```python
plt.figure(figsize=(10,10))
plt.pie(data_pie.age, labels=data_pie.job, startangle=90,
autopct='%.1f%%')
plt.xlabel('AGE')
plt.ylabel('PEKERJAAN')
plt.title('Pekerjaan Paling Banyak')
plt.show()
```
