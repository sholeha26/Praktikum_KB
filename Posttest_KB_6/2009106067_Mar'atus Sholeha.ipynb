import library

```python
from matplotlib import pyplot as plt
import matplotlib
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
```

mengambil data csv yang bernama data.csv

```python
dataset_utama = pd.read_csv("data.csv")
```

membuat objek model dengan jumlah cluster 3

```python
model = KMeans(n_clusters=3)
```

melihat kolom dataset

```python
dataset_utama.columns
```

memilih atribut yang akan digunakan

```python
points = dataset_utama[['tempo', 'instrumentalness']]
```

fit terhadap model

```python
model.fit(points)
```

menentukan labels

```python
labels = model.labels_
```

```python
print(labels)
```

buat dataframe dengan labels, dan artist sebagai kolom

```python
dataKmeans = pd.DataFrame({'labels': labels, 'Artist': dataset_utama.artist})
```

```python
dataset_utama.head(10)
```

buat croostab ct dan menampilkannya

```python
ct = pd.crosstab(dataKmeans["labels"],dataKmeans["Artist"])
print(ct)
```

membuat visualisasi

ambil semua baris pada kolom tertentu

```python
xs = points.iloc[:,0]
ys = points.iloc[:,1]
```

menampung koordinat dari setiap centroid

```python
centroids = model.cluster_centers_
```

```python
centroids_x = centroids[:,0]
centroids_y = centroids[:,1]
```

menampilkan visualisasi Kmeans, dengan xlabel = tempo dan ylabel = instrumentalness.
hasilnya terdapat 3 cluster berbeda, dengan centroid dibagian bawah.

```python
plt.figure(figsize=(10,10))
plt.scatter(xs,ys,alpha=0.5,c=labels)
plt.scatter(centroids_x,centroids_y,marker="D",s=50,label="Centroid")
plt.xlabel("tempo")
plt.ylabel("instrumentalness")
plt.legend()
plt.show()
```

Menentukan Jumlah Cluster(Elbow Method)

```python
ks = range(1, 9)
inertias = []
```

```python
for k in ks:
    model = KMeans(n_clusters=k)
    model.fit(points)
    inertias.append(model.inertia_)
```

```python
plt.figure(figsize=(10,10))
plt.plot(ks, inertias, '-o')
plt.xlabel('Jumlah Cluster, k')
plt.ylabel('inertia')
plt.xticks(ks)
plt.show()
```
